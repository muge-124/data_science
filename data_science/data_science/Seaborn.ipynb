```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("flights")
```

```python
data=data.pivot("month","year","passengers")
#heatmap
sns.heatmap(data=data,annot=True,fmt="d",cmap="YlGnBu")
plt.show()
```

```python
data = sns.load_dataset("flights")

#line plot
sns.lineplot(data=data ,x="year",y="passengers")
plt.show()
```

```python
data = sns.load_dataset("iris")
#swar plot
sns.swarmplot(data=data,x="species",y="petal_length")
plt.show()
```

```python
data = sns.load_dataset("titanic")
#bar plot
sns.countplot(data=data,x="class")
plt.show()
```

```python
data = sns.load_dataset("tips")
#scatter plot
sns.scatterplot(data=data,x="total_bill",y="tip")
plt.show()
```

```python
data = sns.load_dataset("titanic")
#facet grid
g=sns.FacetGrid(data=data,col="sex")
g.map(sns.histplot,"age")
plt.show()
```

```python
data = sns.load_dataset("iris")
g=sns.clustermap(data.drop("species",axis=1),cmap="YlGnBu",standard_scale=1)
plt.show()
```
